```python
# @title üéµ infiniteLoop: Tools Looping Otomatis (Audio & Video)
import os
import subprocess
from google.colab import files
import shutil
import math

# --- 1. BERSIH-BERSIH ---
if os.path.exists('input_file'):
    shutil.rmtree('input_file')
os.makedirs('input_file')

# --- 2. UPLOAD ---
print(">>> UPLOAD FILE (Video MP4 atau Audio MP3/WAV) <<<")
uploaded = files.upload()

if uploaded:
    # Ambil nama file
    filename = list(uploaded.keys())[0]
    filepath = os.path.join('input_file', filename)
    shutil.move(filename, filepath)

    # Deteksi apakah ini Video atau Audio berdasarkan ekstensi
    ext = os.path.splitext(filename)[1].lower()
    is_video = ext in ['.mp4', '.mov', '.avi', '.mkv']

    print(f"\n‚úÖ File terdeteksi: {filename}")

    # --- 3. INPUT DURASI TARGET ---
    # @markdown **Mau dijadikan berapa menit?**
    target_menit = 60 # @param {type:"integer"}

    print(f"‚è≥ Sedang menghitung kebutuhan looping untuk durasi {target_menit} menit...")

    # Cek durasi asli file menggunakan FFprobe
    result = subprocess.run(
        ["ffprobe", "-v", "error", "-show_entries", "format=duration", "-of", "default=noprint_wrappers=1:nokey=1", filepath],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT
    )
    try:
        duration_sec = float(result.stdout)
    except:
        print("‚ùå Gagal membaca durasi file. Pastikan file media valid.")
        duration_sec = 0

    if duration_sec > 0:
        # Hitung butuh berapa kali loop
        target_sec = target_menit * 60
        loop_count = math.ceil(target_sec / duration_sec)

        print(f"   Durasi Asli: {duration_sec:.2f} detik")
        print(f"   Target: {target_sec} detik")
        print(f"   üîÑ File akan di-loop sebanyak: {loop_count} kali")

        # --- 4. PROSES LOOPING (Teknik Concat Demuxer - Super Cepat) ---
        # Kita buat list file teks untuk dibaca FFmpeg (lebih hemat memori daripada load video)
        list_txt = "mylist.txt"
        with open(list_txt, "w") as f:
            for _ in range(loop_count):
                f.write(f"file '{filepath}'\n")

        output_filename = f"LOOPED_{target_menit}min_{filename}"

        print("\nüöÄ Mulai menjahit file (Stitching)...")

        # Perintah FFmpeg (Copy Stream = Tanpa Re-encode = Kualitas Asli & Cepat)
        cmd = f'ffmpeg -f concat -safe 0 -i {list_txt} -c copy "{output_filename}" -y'
        os.system(cmd)

        # --- 5. DOWNLOAD ---
        if os.path.exists(output_filename):
            file_size = os.path.getsize(output_filename) / (1024 * 1024) # dalam MB
            print(f"\nüéâ SELESAI! Ukuran file akhir: {file_size:.2f} MB")
            print("Mengunduh...")
            files.download(output_filename)
        else:
            print("‚ùå Gagal membuat output.")
    else:
        print("‚ùå File kosong atau durasi tidak terbaca.")

else:
    print("‚ùå Kamu tidak mengupload file apa-apa.")
```
